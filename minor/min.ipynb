**Task 4 Data cleaning**


```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.duplicated().sum()
```

```python
df_clean=df.copy()
df_clean["umpire3"]=df_clean["umpire3"].fillna("unknown")
df_clean["city"]=df_clean["city"].fillna("unknown")
df_clean["winner"]=df_clean["winner"].fillna("unknown")
df_clean["umpire1"]=df_clean["umpire1"].fillna("unknown")
df_clean["umpire2"]=df_clean["umpire2"].fillna("unknown")
df_clean["player_of_match"]=df_clean["player_of_match"].fillna("unknown")
```

```python
df_clean.isnull().sum()
```

**5. Outlier Detection**

```python
sns.boxplot(data=df_clean)
```

```python
Q1=df_clean["id"].quantile(0.25)
Q3=df_clean["id"].quantile(0.75)
```

```python
IQR = Q3 - Q1
Lower_Fence = Q1 - 1.5 * IQR
Upper_Fence = Q3 + 1.5 * IQR
```

```python
outliers = df_clean[(df_clean["id"] < Lower_Fence) | (df_clean["id"] > Upper_Fence)]

outliers
```

**11 Conclusion**

In this project, the IPL matches dataset was explored to better understand team performances, match outcomes, and overall tournament trends. The data was first cleaned by handling missing values, removing duplicate records, and converting columns to the appropriate data types. Outliers were identified using boxplots and managed with the IQR method to improve the quality of the analysis.

The exploratory data analysis revealed several interesting patterns. Many matches were won while chasing the target, and most teams preferred to field first after winning the toss. However, winning the toss did not always result in winning the match. The analysis also showed that a few teams have consistently performed well over multiple seasons. Finally, a new Match Type feature (Defended/Chased) was created, which provided additional insights into different winning strategies. Overall, this analysis demonstrates how EDA can uncover valuable patterns and support better decision-making in sports analytics.
