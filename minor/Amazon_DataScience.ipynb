```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

### Q1. Load the dataset and display the first 10 rows.

```python
df = pd.read_csv('amazon.csv')
df.head(10)
```

### Q2. Find the number of rows and columns, and display the data type of each column.

```python
print(df.shape)
print()
print(df.dtypes)
```

### Q3. Find the number of missing values in each column.

```python
print( df.isnull().sum())
```

### Q4. Display all products with a rating greater than 4.5.

```python

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')


print(df[df['rating'] > 4.5]["product_name"])
```

### Q5. Convert the rating column to a NumPy array. Find its mean, minimum, and maximum values.

```python
rating_array = df['rating'].dropna().to_numpy()
print()
print(np.mean(rating_array))
print(np.min(rating_array))
print(np.max(rating_array))
```

### Q6 Find the index of the product with the highest rating using NumPy.

```python

top_product = df.loc[df['rating'].idxmax()]

print(top_product['product_name'])
print(top_product['rating'])
```

### Q7. Plot a histogram of the rating column. Add a title, x-label, and y-label.

```python
plt.figure(figsize=(8, 5))

plt.hist(rating_array, bins=20, color='violet')

plt.title('Product Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Products')

plt.show()
```

### Q8. Create a scatter plot between actual_price (x-axis) and discounted_price (y-axis).

```python

df['actual_price_clean'] = (
    df['actual_price']
    .str.replace('[₹,]', '', regex=True)
    .str.strip()
)
df['actual_price_clean'] = pd.to_numeric(df['actual_price_clean'], errors='coerce')

df['discounted_price_clean'] = (
    df['discounted_price']
    .str.replace('[₹,]', '', regex=True)
    .str.strip()
)
df['discounted_price_clean'] = pd.to_numeric(df['discounted_price_clean'], errors='coerce')


price_df = df.dropna(subset=['actual_price_clean', 'discounted_price_clean'])

plt.figure(figsize=(8, 6))

plt.scatter(
    price_df['actual_price_clean'],
    price_df['discounted_price_clean'],
    color='cyan', alpha=0.78
)

plt.title('Actual Price vs Discounted Price')
plt.xlabel('Actual Price ')
plt.ylabel('Discounted Price ')


plt.show()
```

### Q9. Draw a pie chart showing the share of products in the top 5 main categories.

```python
import matplotlib.pyplot as plt


df['main_category'] = df['category'].str.split('|', regex=False).str[0].str.strip()


top5_pie = df['main_category'].value_counts().head(5)


colors = ['red', 'green', 'blue', 'purple', 'yellow'][:len(top5_pie)]
explode = [0.03] * len(top5_pie)


plt.figure(figsize=(8, 6))

plt.pie(
    top5_pie.values,
    labels=top5_pie.index,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,

)

plt.title('Product Share - Top 5 Categories')
plt.axis('equal')
plt.show()
```

### Q10. Use sns.histplot() to plot the distribution of the rating column with a KDE curve.

```python
plt.figure(figsize=(8, 5))

sns.histplot(
    df['rating'].dropna(),
    bins=20,
    kde=True,
    color='steelblue'
)

plt.title('Rating Distribution with KDE Curve')
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

Q11. Create a sns.countplot() to show the number of products in each of the top 5 main categories.

```python
top5_names = df['main_category'].value_counts().head(5).index.tolist()
df_top5 = df[df['main_category'].isin(top5_names)]
plt.figure(figsize=(9, 5))
sns.countplot(
    data=df_top5,
    x='main_category',
    order=top5_names,
    palette='Blues_d'
)
plt.title('Number of Products — Top 5 Categories')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

### Q12. Create a correlation heatmap using sns.heatmap() for the numerical columns. Annotate each cell with its value.

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['actual_price_clean'] = (
    df['actual_price']
    .astype(str)
    .str.replace('[₹,]', '', regex=True)
    .str.strip()
)
df['actual_price_clean'] = pd.to_numeric(df['actual_price_clean'], errors='coerce')

df['discounted_price_clean'] = (
    df['discounted_price']
    .astype(str)
    .str.replace('[₹,]', '', regex=True)
    .str.strip()
)
df['discounted_price_clean'] = pd.to_numeric(df['discounted_price_clean'], errors='coerce')

num_cols = df[['rating', 'actual_price_clean', 'discounted_price_clean']]

corr_matrix = num_cols.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap — Numerical Columns')
plt.show()
```
