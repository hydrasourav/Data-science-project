Q36. Merge two DataFrames using a common key.

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})
merged_df = pd.merge(df1, df2, on='key')
display(merged_df)
```

Q40. Mini Project: Create a sales DataFrame and answer business questions using filtering, groupby, and aggregation.

## Q40. Mini Project: Create a sales DataFrame and answer business questions using filtering, groupby, and aggregation.

```python
sales_data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03']),
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Units_Sold': [10, 15, 12, 8, 20, 5],
    'Price_Per_Unit': [2.5, 3.0, 2.5, 4.0, 3.0, 2.5]
}
sales_df = pd.DataFrame(sales_data)
sales_df['Total_Sales'] = sales_df['Units_Sold'] * sales_df['Price_Per_Unit']
display(sales_df)
```

### Business Question 1: What are the total sales per region?

```python
total_sales_by_region = sales_df.groupby('Region')['Total_Sales'].sum().reset_index()
display(total_sales_by_region)
```

### Business Question 2: Which product had the highest units sold on '2023-01-01'?

```python
sales_on_jan_1 = sales_df[sales_df['Date'] == '2023-01-01']
highest_units_jan_1 = sales_on_jan_1.loc[sales_on_jan_1['Units_Sold'].idxmax()]
display(highest_units_jan_1)
```

### Business Question 3: What is the average price per unit for each product?

```python
avg_price_per_product = sales_df.groupby('Product')['Price_Per_Unit'].mean().reset_index()
display(avg_price_per_product)
```
